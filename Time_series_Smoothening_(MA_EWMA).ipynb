```python
import pandas as pd
from datetime import datetime
import yfinance as yf
```

```python
df_tesla = yf.download('TSLA')
df_tesla.head()
```

```python
# Simple Moving Average
df_tesla['Open_rolling_10days']=df_tesla['Open'].rolling(window=10).mean()
df_tesla['Open_rolling_20days']=df_tesla['Open'].rolling(window=20).mean()
df_tesla['Open_rolling_30days']=df_tesla['Open'].rolling(window=30).mean()
df_tesla['Open_rolling_40days']=df_tesla['Open'].rolling(window=40).mean()
df_tesla[['Open','Open_rolling_10days','Open_rolling_20days','Open_rolling_30days','Open_rolling_40days']].plot(xlim=['2020-01-01','2021-01-01'],figsize=(12,8))

```

```python
# Cumulative Average or Expanding
df_tesla['Open_expanding']=df_tesla['Open'].expanding().mean()
df_tesla[['Open','Open_expanding']].plot(figsize=(12,8))
```

```python
# Exponential Weighted Moving Average - EWMA
df_tesla['Open_ewma_0.1']=df_tesla['Open'].ewm(alpha=0.1).mean()
df_tesla['Open_ewma_0.5']=df_tesla['Open'].ewm(alpha=0.5).mean()
df_tesla[['Open','Open_ewma_0.1','Open_ewma_0.5']].plot(xlim=['2020-01-01','2021-01-01'],figsize=(12,8))
```
