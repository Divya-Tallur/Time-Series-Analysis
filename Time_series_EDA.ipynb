# **Time Resampling**

```python
# Rule A is year end frequency
# Rule QS is Quarterly start
df_tesla.resample(rule='A').max()['Open'].plot()
```

# **Rolling**

```python
df_tesla.resample(rule='A').max()['Open'].rolling(window=3).mean()
```

```python
df_tesla['Open_rolling_30days']=df_tesla['Open'].rolling(window=30).mean()
```

```python
df_tesla.head(50)
```

```python
df_tesla[['Open','Open_rolling_30days']].plot(xlim=['2020-01-01','2024-01-01'],figsize=(16,8))
```
